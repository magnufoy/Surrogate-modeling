```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
for i in range(1, 101):
    df = pd.read_csv('Simulation_BENDING/CONSTANT_THICKNESS_100/BENDING_X_data.csv/BENDING_'+str(i)+'_data.csv')
    for j in range(0,len(df)): # Find the index of the last zero of the force
        if df.iloc[j]['FKN'] == 0.0:
            index_of_last_zero = j
    df = df.loc[index_of_last_zero:]
    plt.plot(df['DMM'], df['FKN'], linewidth=0.5, color='r')
for i in range(1, 51):
    df = pd.read_csv('Simulation_BENDING/VARIABLE_THICKNESS_50/BENDING_X_data.csv/BENDING_'+str(i)+'_data.csv')
    plt.plot(df['DMM'], df['FKN'], linewidth=0.5, color='b')
for i in range(1,2):
    df = pd.read_csv('Simulation_BENDING/MEAN_VALUES_2/BENDING_X_data.csv/BENDING_'+str(i)+'_data.csv')
    plt.plot(df['DMM'], df['FKN'], linewidth=2.5, color='g')

plt.xlabel('Displacement (mm)')
plt.ylabel('Force (kN)')

# Red: The standard without variable thickness
# Blue: Variable thickness
# Green: Middle value without variable thickness
```
