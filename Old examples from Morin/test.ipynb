```python
import numpy as np
import matplotlib.pyplot as mp
import pandas as pd 
import seaborn as sb 
```

```python
def model(x,y,z):
    f1 = np.cos(x)+y**2.0-z
    f2 = np.sin(x)+y**2.0-z**4.0
    f3 = np.cos(x)+np.exp(-y)
    return f1,f2,f3
```

```python
x = np.random.uniform(low=0.0, high=1.0, size=300)
y = np.random.uniform(low=0.0, high=1.0, size=300)
z = np.random.uniform(low=0.0, high=1.0, size=300)
```

```python
f1,f2,f3 = model(x,y,z)
```

```python
df = pd.DataFrame(data={'X':x,'Y':y,'Z':z,'F1':f1,'F2':f2,'F3':f3})
```

```python
dataplot = sb.heatmap(df.corr(), cmap="Reds", annot=True)  
mp.show() 
```

```python
sb.pairplot(df)
```
