```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**3

# Generate x values
x_values = np.linspace(-5, 5, 100)  # Generate 100 points between -5 and 5

# Calculate corresponding y values using the function
y_values_1 = f(x_values)
y_values_2= f(x_values)+30
y_values_3 =f(x_values)-30

# Plot the function
plt.plot(x_values, y_values_1, label='$f(x) = x^3$', color='blue')
plt.plot(x_values, y_values_2, label='$f(x) = x^3+$', color='blue', linestyle = 'dotted')
plt.plot(x_values, y_values_3, label='$f(x) = x^3-$', color='red',linestyle = 'dotted')
plt.xlabel('x',fontsize =22)
plt.ylabel('f(x)', fontsize =22)
plt.xticks(fontsize=20)  
plt.yticks(fontsize=20)
plt.grid(True)
plt.legend(fontsize = 16)
plt.show()
```
