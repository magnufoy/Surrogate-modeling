```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python


# Read the CSV file into a pandas DataFrame
data = pd.read_csv('Energy_data_plot.csv', sep=';', header=None)
data = data.astype(float)
# Extract the data from the first two columns
x_data_ALLAE = data.iloc[:201, 0]
y_data_ALLAE = data.iloc[:201, 1]

# Extract the title of x and y axis from the first row
x_data_ALLIE = data.iloc[1:201, 2]
y_data_ALLIE = data.iloc[1:201, 3]
# Plot the data
plt.figure(dpi=300)
plt.plot(x_data_ALLAE, y_data_ALLAE, linewidth= 2.5, color = 'tab:blue')
plt.plot(x_data_ALLIE, y_data_ALLIE, linewidth= 2.5, color = 'tab:orange')
plt.xlabel('Time (s)',fontsize=22)
plt.ylabel('Energy (J)', fontsize=22)

plt.yticks(fontsize=22)
plt.xticks(fontsize=22)
plt.grid(True)
plt.show()
```

```python

# Read the CSV file into a pandas DataFrame
data = pd.read_csv('Energy_data_plot.csv', sep=';', header=None)
data = data.astype(float)
# Extract the data from the first two columns
x_data_ALLAE = data.iloc[:201, 4]
y_data_ALLAE = data.iloc[:201, 5]

# Extract the title of x and y axis from the first row
x_data_ALLIE = data.iloc[1:201, 6]
y_data_ALLIE = data.iloc[1:201, 7]
# Plot the data
plt.figure(dpi=300)
plt.plot(x_data_ALLAE, y_data_ALLAE, linewidth= 2.5, color = 'tab:blue')
plt.plot(x_data_ALLIE, y_data_ALLIE, linewidth= 2.5, color = 'tab:orange')
plt.yticks(fontsize=22)
plt.xticks(fontsize=22)
# Get the current axes
ax = plt.gca()

# Set the offset font size
ax.yaxis.get_offset_text().set_fontsize(22)  # Change the font size as needed
plt.xlabel('Time (s)',fontsize=22)
plt.ylabel('Energy (J)', fontsize=22)
plt.grid(True)
plt.show()
```
